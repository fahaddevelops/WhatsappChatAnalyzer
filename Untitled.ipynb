```python
import re
import pandas as pd
import nltk
import matplotlib.pyplot as plt
# from collection import Counter
```

```python
f = open('chats.txt', 'r', encoding = 'utf-8')
```

```python
data = f.read()
pattern = r'(\d{1,2}/\d{1,2}/\d{4},\s\d{1,2}:\d{2}\s[APapMm]{2}\s-\s)(.*)'
```

```python
msgs = re.findall(pattern, data)
user_msg = []
msg_date = []

for date, msg in msgs:
    user_msg.append(msg)
    msg_date.append(date.strip())  # Remove extra spaces v 

```

```python
df = pd.DataFrame({'user_msg': user_msg, 'msg_date': msg_date})
# convert msgs date type
```

```python
df['msg_date'] = pd.to_datetime(df['msg_date'], format='%d/%m/%Y, %I:%M %p -')

df.rename(columns = {'msg_date': 'date'}, inplace = True)
df.head()
```

```python
df.shape
```

```python
users = []
messages = []

for msg in df['user_msg']:  
    entry = re.split('([\w\W]+?):\s', msg)
    if entry[1:]:
        users.append(entry[1])
        messages.append(entry[2])
    else:
        users.append('Group_notification')
        messages.append(entry[0])

df['user'] = users
df['message'] = messages

df.drop(columns=['user_msg'], inplace=True)

df.head(15)

```

```python
stop_words = set(stopwords.words('english'))
urdu_stop_words = {'k', 'ap', 'ok', 'ma', 'ko', 'ha', 'ki', 'kr', 'ho', 'hi', 'bhi', 'b', 'na', 'ka', 'ni' }  # Add more Urdu stopwords as needed
all_stop_words = stop_words.union(urdu_stop_words) 
```
